```python

import pandas as pd

import matplotlib.pyplot as plt
# Actual values for each year from 2018 to 2022
actual_values = {
    '2018': [10.0, 19.0, 12.0, 6.0, 11.0, 6.0, 6.0, 8.0, 13.0, 8.0],
    '2019': [8.0, 17.0, 10.0, 6.0, 16.0, 8.0, 3.0, 8.0, 13.0, 10.0],
    '2020': [17.7, 23.0, 8.7, 11.1, 10.2, 4.0, 6.6, 7.9, 5.1, 5.7],
    '2021': [23.2, 22.4, 12.7, 8.2, 7.3, 2.8, 5.1, 2.8, 4.0, 2.5],
    '2022': [24.8, 18.9, 14.6, 10.7, 8.7, 7.3, 5.8, 4.8, 3.9, 0.5]
}

# Forecasted values for 2023
forecasted_values_2023 = [10.0, 19.0, 12.0, 6.0, 11.0, 6.0, 6.0, 8.0, 13.0, 8.0]

# Calculate Mean Absolute Error (MAE) for each year
mae_values = []
for year in actual_values:
    n = len(actual_values[year])
    mae = sum(abs(actual - forecast) for actual, forecast in zip(actual_values[year], forecasted_values_2023)) / n
    mae_values.append(mae)

# Calculate the average MAE for the entire period
average_mae = sum(mae_values) / len(mae_values)

print("Mean Absolute Error (MAE) for each year:", mae_values)
print("Average MAE for the entire period:", average_mae)
```

```python
# Example data for visualization
forecast_periods = ['2018', '2019', '2020', '2021', '2022', '2023']
mae_values = [0.0, 1.8, 3.38, 4.7, 4.58, 2.892]

# Create a pandas DataFrame
data = {'Forecast Period': forecast_periods, 'MAE': mae_values}
df = pd.DataFrame(data)

# Create a line plot
plt.plot(df['Forecast Period'], df['MAE'], marker='o')
plt.xlabel('Forecast Period')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('MAE over Different Forecast Periods')
plt.grid(True)
plt.show()
```
